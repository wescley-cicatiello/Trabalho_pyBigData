Dados das viagens a serviço do Governo Federal oriundos do Sistema de Concessão de Diárias e Passagens (SCDP).

O SCDP é a solução responsável pela gestão dos processos de afastamento a serviço do Governo Federal. Permite a elaboração, o registro, o controle, o acompanhamento e a gestão dos processos de concessão de diárias e passagens.

O Sistema possibilita, ainda, o compartilhamento de uma base de dados única, administrada pelo Gestor Central no Ministério da Economia (ME) e pelos Gestores Setoriais de cada instituição usuária, o que permite um maior controle físico e financeiro das diárias e passagens emitidas no âmbito da Administração Pública Federal.

Foi escolhida essa fonte de dados por serem dados reais e em uma quantidade suficiente para executar varias formas de analise. Assim como a disponibilidade já de um dicionário de dados.

Neste primeiro exemplo iremos agrupar os Orgãos e a quantidade de passagens.

```python
df = pd.read_excel("202406_CompraDireta_EmissoesAlteracoes.ods", engine="odf")

# Agrupa e conta a quantidade de passagens por órgão
qtdByOrg = df.groupby("nome_orgao_superior").size().reset_index(name="quantidade_passagens")

# Exibe o resultado
print(qtdByOrg)
```

Agora iremos agrupar a quantidade de passagens por empresa aérea

```python
# Lê o arquivo .ods
df = pd.read_excel("202406_CompraDireta_EmissoesAlteracoes.ods", engine="odf")

# Agrupa por companhia aérea e conta as passagens
passagens_por_companhia = df.groupby("companhia_aerea").size().reset_index(name="quantidade_passagens")

# Ordena do maior para o menor
passagens_por_companhia = passagens_por_companhia.sort_values(by="quantidade_passagens", ascending=False)

# Exibe o resultado
print(passagens_por_companhia)
```

Vamos demonstrar o valor gasto por cada Orgão

```python
df = pd.read_excel("202406_CompraDireta_EmissoesAlteracoes.ods", engine="odf")

#Garante que a coluna valor_bilhete é numérica
df["valor_bilhete"] = pd.to_numeric(df["valor_bilhete"], errors="coerce")

# agrupa por orgão e soma os valores dos bilhetes
soma_bilhetes = df.groupby("nome_orgao_superior")["valor_bilhete"].sum().reset_index()

#renomeia a coluna de resultado
soma_bilhetes = soma_bilhetes.rename(columns={"valor_bilhete": "total_gasto_bilhete"})

# Formata os valores como R$ 1.234,56
soma_bilhetes["total_gasto_bilhete"] = soma_bilhetes["total_gasto_bilhete"].apply(
    lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
)

#exibe resultado
print(soma_bilhetes)
```

E tambem o valor separado por cada empresa aérea

```python
# Lê o arquivo .ods
df = pd.read_excel("202406_CompraDireta_EmissoesAlteracoes.ods", engine="odf")

# Garante que a coluna 'valor_bilhete' seja numérica
df["valor_bilhete"] = pd.to_numeric(df["valor_bilhete"], errors="coerce")

# Agrupa por companhia aérea e soma os valores dos bilhetes
soma_por_companhia = df.groupby("companhia_aerea")["valor_bilhete"].sum().reset_index()

# Renomeia a coluna com o resultado
soma_por_companhia = soma_por_companhia.rename(columns={"valor_bilhete": "total_gasto_bilhetes"})

# Formata os valores como R$ 1.234,56
soma_por_companhia["total_gasto_bilhetes"] = soma_por_companhia["total_gasto_bilhetes"].apply(
    lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
)

# Ordena do maior para o menor
soma_por_companhia = soma_por_companhia.sort_values(by="total_gasto_bilhetes", ascending=False)

# Exibe o resultado
print(soma_por_companhia)
```

Agora podemos ver a quantidade de voos por data.

```python
# Nome do arquivo .ods
arquivo_ods = "202406_CompraDireta_EmissoesAlteracoes.ods"

# Lê o arquivo usando o engine odf
df = pd.read_excel(arquivo_ods, engine="odf")

# Verifica se a coluna 'data_embarque' existe
if "data_embarque" not in df.columns:
    print("A coluna 'data_embarque' não foi encontrada no arquivo.")
else:
    # Converte a coluna para data (caso ainda não esteja)
    df["data_embarque"] = pd.to_datetime(df["data_embarque"], errors="coerce")

    # Remove valores inválidos ou vazios
    df = df.dropna(subset=["data_embarque"])

    # Agrupa por data e conta os embarques
    qtd_por_data = df.groupby(df["data_embarque"].dt.date).size().reset_index(name="quantidade_embarques")

    # Ordena por data (opcional)
    qtd_por_data = qtd_por_data.sort_values(by="data_embarque")

    # Exibe o resultado
    print(qtd_por_data)
```

Outra forma de apresentar para que fique mais facil de observar.

```python
# Gráfico de barras verticais
plt.figure(figsize=(14, 6))
plt.bar(qtd_por_data["data_embarque"].astype(str), qtd_por_data["quantidade_embarques"], color="skyblue")
plt.title("Quantidade de Embarques por Data")
plt.xlabel("Data do Embarque")
plt.ylabel("Quantidade de Embarques")
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

Agora vamos ver o valor gasto por dia.

```python
arquivo_ods = "202406_CompraDireta_EmissoesAlteracoes.ods"

df = pd.read_excel(arquivo_ods, engine="odf")

# Verifica se as colunas necessárias existem
if "data_embarque" not in df.columns or "valor_bilhete" not in df.columns:
    print("As colunas 'data_embarque' e/ou 'valor_bilhete' não foram encontradas.")
else:
    # Converte a coluna de datas e valores
    df["data_embarque"] = pd.to_datetime(df["data_embarque"], errors="coerce")
    df["valor_bilhete"] = pd.to_numeric(df["valor_bilhete"], errors="coerce")

    # Remove linhas com dados inválidos
    df = df.dropna(subset=["data_embarque", "valor_bilhete"])

    # Agrupa por data e soma os valores
    gastos_por_data = df.groupby(df["data_embarque"].dt.date)["valor_bilhete"].sum().reset_index(name="total_gasto")

    # Ordena por data
    gastos_por_data = gastos_por_data.sort_values(by="data_embarque")

    # Formata os valores como R$ 1.234,56
    gastos_por_data["total_gasto_formatado"] = gastos_por_data["total_gasto"].apply(
        lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    )

    # Exibe a tabela com valores formatados
    print(gastos_por_data[["data_embarque", "total_gasto_formatado"]])
```

Grafico para facilitar apresentação

```python
# (Opcional) Gráfico de barras para visualizar os gastos por dia
plt.figure(figsize=(14, 6))
plt.bar(gastos_por_data["data_embarque"].astype(str), gastos_por_data["total_gasto"], color="steelblue")
plt.title("Total Gasto por Dia de Embarque")
plt.xlabel("Data")
plt.ylabel("Valor Gasto (R$)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```
